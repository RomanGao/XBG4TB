# 不同药物的实验数据分布

```python
msf_subcol = pd.read_csv("../1_data/msf_subcol.csv")
msf_subcol
```

```python
from collections import Counter
for name, group in msf_subcol.groupby("MSDRUG"):
    print(name,":")
    ct = Counter(group["MSSTRESC"])
    print(ct)
```

```python
xx = []
ys = []
yr = []

for name, group in msf_subcol.groupby("MSDRUG"):
    print(name)
    ct = Counter(group["MSSTRESC"])

    xx.append(name)
    ys.append(ct['SUSCEPTIBLE'])
    yr.append(ct['RESISTANT'])
plt.figure(figsize=(12,9))
plt.barh(xx, ys,  color="b", alpha=0.3,label='S')
plt.barh(xx, yr, color='r' ,left= ys,label='R')
plt.legend(loc='lower right')
plt.savefig("../2_data/drug_sr_ct.png")
plt.show()
```

#### 真正代码

--------》》》

```python
for f in os.listdir('../2_data/gene/single/'):
    tmp_df = pd.read_csv('../2_data/gene/single/'+f)
    feat = list(set(tmp_df.GENENAME))
#     feat.insert(0, 'ISOLATID')
    feat.append('label')
    df_save = pd.DataFrame(columns= feat)
    
    aa = 0
    for i, group in tmp_df.groupby('ISOLATID'):
        gct = Counter(group.GENENAME.values)
#         gct['ISOLATID'] = i
        gct['label'] = group.label.values[0]
        df_save = df_save.append(pd.Series(gct), ignore_index=True)
        df_save = df_save.fillna(0).astype(int)
        aa+=1
        if(aa==3):
            break
#     print(df_save.head)
    df_save.to_csv("../2_data/gene/single_feature/"+f+".csv", index=False)
```
