```python
from collections import Counter
for name, group in msf_subcol.groupby("MSDRUG"):
    print(name,":")
    ct = Counter(group["MSSTRESC"])
    print(ct)
```

```python
xx = []
ys = []
yr = []

for name, group in msf_subcol.groupby("MSDRUG"):
    print(name)
    ct = Counter(group["MSSTRESC"])

    xx.append(name)
    ys.append(ct['SUSCEPTIBLE'])
    yr.append(ct['RESISTANT'])
plt.figure(figsize=(12,9))
plt.barh(xx, ys,  color="b", alpha=0.3,label='S')
plt.barh(xx, yr, color='r' ,left= ys,label='R')
plt.legend(loc='lower right')
plt.savefig("../2_data/drug_sr_ct.png")
plt.show()
```

### 生成MDR代码

```python
df_multi = pd.DataFrame(columns= feat)
```

```python
a = 0
df_multi = pd.DataFrame(columns= feat)
for ii, group in pff_gene.groupby('ISOLATID'):
    if set(first_Drug).issubset(set(group.MSDRUG.values)):
        a+=1
        gennu = list(set(group.GENE_NU.values))
        gg = [ii.split("_")[0] for ii in gennu]
        gct = collections.Counter(gg)
        gct['ISONIAZID'] = group[group.MSDRUG == 'ISONIAZID'].label.values[0]
        gct['RIFAMPICIN'] = group[group.MSDRUG == 'RIFAMPICIN'].label.values[0]
        gct['ETHAMBUTOL'] = group[group.MSDRUG == 'ETHAMBUTOL'].label.values[0]
        gct['PYRAZINAMIDE'] = group[group.MSDRUG == 'PYRAZINAMIDE'].label.values[0]
        df_multi = df_multi.append(pd.Series(gct), ignore_index= True)
        df_multi = df_multi.fillna(0).astype(int)

#     print(df_multi.head)
        print(a)
df_multi.to_csv("../2_data/gene/multi_feature/MDRv2.csv", index=False)
```
