```python
# finding the threshold based on probability and labe;s of the validation set    
def find_thr(pred,label):
    
    #find the best threshold where false possitive  rate and falsi negative points cross
    minn=100000
    thrr=0.4
    
    for thr in np.arange(0.1,1,0.1):
        prr = np.where(pred > thr, 1, 0)
#         print(label,",,,",prr)
        tn, fp, fn, tp = confusion_matrix(label,prr).ravel()
        if tp+fn > 0:
            frr=fn/(tp+fn)
        else:
            frr = 0
        if tn+fp > 0:    
            far=fp/(tn+fp)
        else:
            far = 0 
        if np.abs(frr - far) < minn:
            minn=np.abs(frr - far)
            thrr=thr
            
    return thrr

#subsample equally for imbalace dataset by selecting samples randomly
#x and y are the data and the corresponding labels
def balanced_subsample(x,y):
    class_xs = []
    min_elems = None
    subsample_size=1.0 # keep equal number of samples for all classes
    
    for yi in np.unique(y): # for any class in our data
        elems = x[(y == yi)] #find the data of each label
        class_xs.append((yi, elems)) 
        if min_elems == None or elems.shape[0] < min_elems: # find min_elemns as the group with lower samples
            min_elems = elems.shape[0]

    use_elems = min_elems # if proportion is not 1 not applicable in our work
    if subsample_size < 1:
        use_elems = int(min_elems*subsample_size)

    xs = []
    ys = []

    for ci,this_xs in class_xs: # randomly select samples from larger group
        if len(this_xs) > use_elems:
            np.random.shuffle(this_xs)

        x_ = this_xs[:use_elems]
        y_ = np.empty(use_elems)
        y_.fill(ci)
        
        xs.append(x_)
        ys.append(y_)
        
    # concatenate the data of all groups before returning them 
    xs = np.concatenate(xs)
    ys = np.concatenate(ys)
    
    return xs,ys

#calculate the performance (Accuracy, sensitivity, specificity, AUC)
def performance_calculation(array1,array2,array3):
#     print(array1, "||| ", array2)
    tn, fp, fn, tp = confusion_matrix(array1,array2).ravel()
    total=tn+fp+fn+tp
    acc= (tn+tp)/total
    sen = tp/(tp+fn)
    sps = tn/(tn+fp)

    fpr, tpr, thresholds = metrics.roc_curve(array1, array3)
    roc_auc=metrics.auc(fpr, tpr)

    return acc,sen,sps,roc_auc 
```

```python
alexander = -1
# Samaneh Kouchaki
# This file includes multi and single label RF classifiers
#In all classifiers we run 10 runs of 5-fold crossvalidation in each run 20 percent keeps as test set and 80 percent train and validation sets
#for setting the classifier threshold (it is based on best accuracy on validation set)
#10 fold cross validation and 5 runs

class EnsembleLearning_multilabel:
    #load data and label
    #filename1 is the feature matrix (rows:isolates and cols: features) and filename2 is labels one for each isolates
    def __init__(self, feat_df,label_df):
        self.feat=feat_df.values
        self.label=label_df.values 
        
    #multi-label RF
    # name1: all mutations list and name2: important features list for all drugs 
    #name1 and name2 are for loading important features. It is used only if not none otherwise the whole feature space is used
    def rf_multilabel(self,path = ""):
        #keeping performance
        # 10 runs of 5 fold (50) for EMB, INH, RIF, PZA, XDR, MDR (6)
        accm=np.zeros((25,4))
        senm=np.zeros((25,4))
        spsm=np.zeros((25,4))
        aucm=np.zeros((25,4))
        kk = 0
        
        wdata = self.feat
        wlabel = self.label
        wlabel = wlabel.astype(int)
        
 
        for i in range(5):
            model = RandomForestClassifier(n_estimators=50, max_depth=None,max_features= int(np.floor(np.sqrt(len(wdata[0]))))) #rf
#             rf = RandomForestClassifier()
#             model = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = alexander)
            test_fold = np.zeros(len(wlabel))  
            #use iterative split
#           ind = iterative_split(wlabel, 5)
            
            nfold = 5
            kf = MultilabelStratifiedKFold(n_splits=nfold, shuffle=True, random_state=2021)
            for trind, teind in kf.split(wdata, wlabel):
                # 80% as train and validation
                tr=wdata[trind]
                trl=wlabel[trind]
                
                #20% as test
                te=wdata[teind]
                tel=wlabel[teind]
             
                # train validation split based on iterative CV
#                 tind = iterative_split(trl, 5)
                tind = proba_mass_split(trl, 5)
    
                test_label = trl[tind[0]]  
                test=tr[tind[0]]
                tind=np.concatenate(tind[1:5],axis=0)
                train_label=trl[tind]
                train=tr[tind]
                
                # fit the model 
                model.fit(train,train_label)
                
                # find treshold for each label
                pred=model.predict_proba(test)
                thr=[]
                for j in range(len(pred)):
                    thr.append(find_thr(pred[j][:,1],test_label[:,j]))
                pred=model.predict_proba(te)
                prr=np.zeros((len(tel),len(tel[0])))
                
                for j in range(len(pred)):
                    prr[:,j]= np.where(pred[j][:,1] > thr[j], 1, 0)
                
                #accuracy per label
                for j in range(len(prr[0])):
                    acc,sen,sps,auc=performance_calculation(tel[:,j],prr[:,j],prr[:,j])
                    accm[kk][j]=acc
                    senm[kk][j]=sen
                    spsm[kk][j]=sps
                    aucm[kk][j]=auc
                kk=kk+1
                print(kk)
        f = open(path, "w")
        f.write("rf_res ={'accm': np.array("+str((accm*100).tolist())+"),\n'senm':np.array(" + str((senm*100).tolist())+"),\n'spsm': np.array("+str((spsm*100).tolist())+ "),\n'aucm' : np.array("+str((aucm*100).tolist())+")}\n")
        f.close()
        print("accm: ", accm*100, "\nsenm: ", senm*100, "\nspsm: ",spsm*100, "\naucm:", aucm*100)
        return accm *100, senm*100, spsm*100, aucm*100
    
    def xgb_multilabel(self, path = ""):
        #keeping performance
        # 10 runs of 5 fold (50) for EMB, INH, RIF, PZA, XDR, MDR (6)
        accm=np.zeros((25,4))
        senm=np.zeros((25,4))
        spsm=np.zeros((25,4))
        aucm=np.zeros((25,4))
        kk = 0
        
        wdata = self.feat
        wlabel = self.label
        wlabel = wlabel.astype(int)
        

        for i in range(5):
            model = OneVsRestClassifier(XGBClassifier(subsample= 0.9, n_estimators=100, min_child_weight= 1, max_depth= 5, learning_rate= 0.05, gamma=0.05, eta= 0.4, colsample_bytree=0.8,tree_method='gpu_hist'))
            test_fold = np.zeros(len(wlabel))  
            
            
#             kf = KFold(n_splits=5)
#             kf.get_n_splits(wdata, wlabel)
            # 5折
            nfold = 5
            kf = MultilabelStratifiedKFold(n_splits=nfold, shuffle=True, random_state=2021)
            for trind, teind in kf.split(wdata, wlabel):
                # 80% as train and validation
                train=wdata[trind]
                train_label=wlabel[trind]
       
                #20% as test
                test=wdata[teind]
                test_label=wlabel[teind]
             
                # fit the model 
                model.fit(train,train_label)
#                 model = xgb.train(params, xgb.DMatrix(train, train_label))
                
                # find treshold for each label
                pred=model.predict(test)
#                 print(model.predict_proba(test))
                #accuracy per label
                for j in range(len(pred[0])):
                    acc,sen,sps,auc=performance_calculation(test_label[:,j],pred[:,j],pred[:,j])
                    accm[kk][j]=acc
                    senm[kk][j]=sen
                    spsm[kk][j]=sps
                    aucm[kk][j]=auc
                kk=kk+1
                print("--kk--", kk)
#                 break
        # return 
        f = open(path, "w")
        f.write("xgb_res ={'accm': np.array("+str((accm*100).tolist())+"),\n'senm':np.array(" + str((senm*100).tolist())+"),\n'spsm': np.array("+str((spsm*100).tolist())+ "),\n'aucm' : np.array("+str((aucm*100).tolist())+")}\n")
        f.close()
#         print("accm: ", accm*100, "\nsenm: ", senm*100, "\nspsm: ",spsm*100, "\naucm:", aucm*100)
        return accm *100, senm*100, spsm*100, aucm*100
 
    #direct association it does not have any training stage and only workss on test set
    def DA(self):
        #keeping performance
        # 10 runs of 5 fold (50) for EMB, INH, RIF, PZA, XDR, MDR (6)
        accm=np.zeros((50,6))
        senm=np.zeros((50,6))
        spsm=np.zeros((50,6))
        aucm=np.zeros((50,6))
        
        wdata = self.feat
        wlabel = self.label
        wlabel=wlabel.astype(int)
        
        kk = 0
        for i in range(10):
            #iterative split to have fair comparison with machine learning methods
            test_fold = np.zeros(len(wlabel))    
            ind = iterative_split(wlabel, 5)
            for j in range(len(ind)):
                for k in range(len(ind[j])):
                    test_fold[ind[j][k]] = j
            ps = PredefinedSplit(test_fold)
            
            for trind, teind in ps.split():
                #data is the estimated value and ;abel is the true label
                eLabel = wdata[teind]
                tLabel = wlabel[teind]
                
                # performance calculation for each label (drug)
                for j in range(len(eLabel[0])):
                    acc,sen,sps,auc = performance_calculation(tLabel[:,j],eLabel[:,j],eLabel[:,j])
                    accm[kk][j]=acc
                    senm[kk][j]=sen
                    spsm[kk][j]=sps
                    aucm[kk][j]=auc
                kk=kk+1
                print(kk)
        # return 
        return accm * 100, senm*100, spsm*100, aucm*100 
        
    # find indeces of important features to be considered for classification based on important features
    def find_features_multi(self, array1, array2):
        res = []
        x=array1.ravel().tolist()
        try:
            for i in range(len(array2)):
                if pd.isnull(array2[i]) == False:
                    ind = x.index(array2[i])
                    res.append(ind)
        except ValueError as e:
            print(e)
        return res
```

```python
### 不同feature importance系数的特征选择
from collections import Counter
class mutation_ranking:
    #load data and label
    #filename1 is the feature matrix (rows:isolates and cols: features) and filename2 is labels one for each isolates
    def __init__(self, feat,label):
        self.feat=feat
#         self.label = pd.read_csv(filename2,header=None)
        self.label=label


    #rank mutations based on multilabel learning or single label learnng
    # defualt is multi
    #fname contains the mutation list
    def mutation_importance(self,col,path = "", importance_type = None ): 
        
        wdata = self.feat
        wlabel = self.label
        
        
        model = OneVsRestClassifier(XGBClassifier(tree_method='gpu_hist'))
        model.fit(wdata,wlabel)
        print("Feature ranking ...")
#         print(wdata.head())
        feat_impts = [] 
        for clf in model.estimators_:
            feat_impts.append(clf.get_booster().get_score(importance_type=importance_type))
#             print(clf.get_booster().get_score(importance_type=importance_type))
#         print(feat_impts)
        res = Counter()
        for ct in feat_impts:
            res = Counter(ct)+res
        res = dict(res)
#         print(res)
        importance = sorted(res.items(), key = lambda x:x[1], reverse= True)

        imp_df = pd.DataFrame(importance, columns=["feature", importance_type])
#         print(imp_df)
        imp_df[importance_type] = imp_df[importance_type]/imp_df[importance_type].sum()
        imp_df.to_csv(path, index=False)
        print("feature stop ...")
```

```python
from sklearn.feature_selection import SelectFromModel

def select_thred(X_train,y_train,X_test,y_test ,importances=None, path =None):

     
    model = OneVsRestClassifier(XGBClassifier(tree_method='gpu_hist'))
    model.fit(X_train,y_train)
    
    thresholds = []
    for importance in importances:
        if importance not in thresholds:
            thresholds.append(importance)

    thresholds = sorted(thresholds)
    df = pd.DataFrame(columns = ['thred', 'auc'])
    for threshold in thresholds:
        #进行特征选择
#         for clf in model.estimators_:
        selection = SelectFromModel(model.estimators_[0], threshold  = threshold, prefit = True)
        select_x_train = selection.transform(X_train)
        aucm=np.zeros((4,4))
        kk=0
        #训练模型
        selection_model = OneVsRestClassifier(XGBClassifier(tree_method='gpu_hist'))

        selection_model.fit(select_x_train, y_train)

        #评估模型
        select_x_test = selection.transform(X_test)
        pred = selection_model.predict(select_x_test)
    #     auc = roc_auc_score(y_test, y_pred)
    #     print("阈值： %.5f , 特征数量为：%d, AUC 得分： %.2f" %(threshold,select_x_train.shape[1], auc))
        for j in range(len(pred[0])):
            
            aucm[kk][j]=perf_cal(y_test[:,j],pred[:,j])
            kk=kk+1
            print("--kk--", kk)
    #                 break
            # return 
    #     f = open(path, "w")
    #     f.write("xgb_res ={'accm': np.array("+str((accm*100).tolist())+"),\n'senm':np.array(" + str((senm*100).tolist())+"),\n'spsm': np.array("+str((spsm*100).tolist())+ "),\n'aucm' : np.array("+str((aucm*100).tolist())+")}\n")
    #     f.close()
    #         print("accm: ", accm*100, "\nsenm: ", senm*100, "\nspsm: ",spsm*100, "\naucm:", aucm*100)a
        auc = np.array((aucm*100).tolist()).max(axis = 1).mean().round(2)
        
        df = df.append(pd.Series({"thred":threshold, "auc":auc}),ignore_index=True)
    df.to_csv(path, index =False)
```

```python
import eli5
from eli5.sklearn import PermutationImportance
from collections import Counter

class mutation_permutation:
    #load data and label
    #filename1 is the feature matrix (rows:isolates and cols: features) and filename2 is labels one for each isolates
    def __init__(self, feat,label):
        self.feat=feat
#         self.label = pd.read_csv(filename2,header=None)
        self.label=label


    #rank mutations based on multilabel learning or single label learnng
    # defualt is multi
    #fname contains the mutation list
    def mutation_importance(self,col,path = "", importance_type = None ): 
        
        wdata = self.feat
        wlabel = self.label
        
        
        model = OneVsRestClassifier(XGBClassifier(tree_method='gpu_hist'))
        model.fit(wdata,wlabel)
        print("Feature ranking ...")

        perm = PermutationImportance(model, random_state=1).fit(wdata,wlabel)
        pi_features = eli5.explain_weights_df(perm, feature_names = col)
        pi_features.to_csv(path, index=False)
        
        print("feature stop ...")
```

```python
from collections import Counter
class mutation_ranking:
    #load data and label
    #filename1 is the feature matrix (rows:isolates and cols: features) and filename2 is labels one for each isolates
    def __init__(self, feat,label):
        self.feat=feat
#         self.label = pd.read_csv(filename2,header=None)
        self.label=label


    #rank mutations based on multilabel learning or single label learnng
    # defualt is multi
    #fname contains the mutation list
    def mutation_importance(self,col,path = "", importance_type = None ): 
        
        wdata = self.feat
        wlabel = self.label
        
        
        model = OneVsRestClassifier(XGBClassifier(tree_method='gpu_hist'))
        model.fit(wdata,wlabel)
        print("Feature ranking ...")
#         print(wdata.head())
        feat_impts = [] 
        for clf in model.estimators_:
            feat_impts.append(clf.get_booster().get_score(importance_type=importance_type))
#             print(clf.get_booster().get_score(importance_type=importance_type))
#         print(feat_impts)
        res = Counter()
        for ct in feat_impts:
            res = Counter(ct)+res
        res = dict(res)
#         print(res)
        importance = sorted(res.items(), key = lambda x:x[1], reverse= True)

        imp_df = pd.DataFrame(importance, columns=["feature", importance_type])
#         print(imp_df)
        imp_df[importance_type] = imp_df[importance_type]/imp_df[importance_type].sum()
        imp_df.to_csv(path, index=False)
        print("feature stop ...")
```

```python
import warnings


warnings.filterwarnings('ignore') 
alexander = -1
# Samaneh Kouchaki
# This file includes multi and single label RF classifiers
#In all classifiers we run 10 runs of 5-fold crossvalidation in each run 20 percent keeps as test set and 80 percent train and validation sets
#for setting the classifier threshold (it is based on best accuracy on validation set)
#10 fold cross validation and 5 runs

class EnsembleLearning_multilabel:
    #load data and label
    #filename1 is the feature matrix (rows:isolates and cols: features) and filename2 is labels one for each isolates
    def __init__(self, feat_df,label_df):
        self.cols = feat_df.columns
        self.feat=feat_df
        self.label=label_df.values 
        
    #multi-label RF
    # muts : importance feature  
    #muts is none : all feature
    #name1 and name2 are for loading important features. It is used only if not none otherwise the whole feature space is used
    def ensemble_multilabel(self,muts = None):
        #keeping performance
        # 10 runs of 5 fold (50) for EMB, INH, RIF, PZA, XDR, MDR (6)
        accm=np.zeros((25,4))
        senm=np.zeros((25,4))
        spsm=np.zeros((25,4))
        aucm=np.zeros((25,4))
        kk = 0
         #load important features to be used for training
        if muts is not None:
            # it load
            self.cols = muts.name.values[:20].tolist()
            print("importance feature...")
#         print(type(self.feat))
        wdata = self.feat[self.cols].values
        wlabel = self.label
        wlabel = wlabel.astype(int)

        inds =[]
       

        
        for i in range(5):
            model = OneVsRestClassifier(XGBClassifier())
#             rf = RandomForestClassifier()
#             model = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 100, cv = 3, verbose=2, random_state=42, n_jobs = alexander)
            test_fold = np.zeros(len(wlabel))  
            #use iterative split
#             ind = iterative_split(wlabel, 5)
            ind = proba_mass_split(wlabel, 5)
            
            #split data based on iterative method
            for j in range(len(ind)):
                for k in range(len(ind[j])):
                    test_fold[ind[j][k]]=j
            ps = PredefinedSplit(test_fold)
            
            for trind, teind in ps.split():
                # 80% as train and validation
                tr=wdata[trind]
                trl=wlabel[trind]
                
                #20% as test
                te=wdata[teind]
                tel=wlabel[teind]
             
                # train validation split based on iterative CV
#                 tind = iterative_split(trl, 5)
                tind = proba_mass_split(trl, 5)
    
                test_label = trl[tind[0]]  
                test=tr[tind[0]]
                tind=np.concatenate(tind[1:5],axis=0)
                train_label=trl[tind]
                train=tr[tind]
                
                # fit the model 
                model.fit(train,train_label)
                
                # find treshold for each label
                pred=model.predict_proba(test)
                thr=[]
                for j in range(len(pred[0])):
                    thr.append(find_thr(pred[:,j],test_label[:,j]))
                pred=model.predict_proba(te)
                prr=np.zeros((len(tel),len(tel[0])))
                
                for j in range(len(pred[0])):
                    prr[:,j]= np.where(pred[:,j] > thr[j], 1, 0)
                
                #accuracy per label
                for j in range(len(prr[0])):
                    acc,sen,sps,auc=performance_calculation(tel[:,j],prr[:,j],prr[:,j])
                    accm[kk][j]=acc
                    senm[kk][j]=sen
                    spsm[kk][j]=sps
                    aucm[kk][j]=auc
                kk=kk+1
                print(kk)
        # return
        f = open("./dataset_pff/data_multi/xgb_20.txt", "w")
        f.write("xgb_res ={'accm': np.array("+str((accm*100).tolist())+"),\n'senm':np.array(" + str((senm*100).tolist())+"),\n'spsm': np.array("+str((spsm*100).tolist())+ "),\n'aucm' : np.array("+str((aucm*100).tolist())+")}\n")
        f.close()
        return accm *100, senm*100, spsm*100, aucm*100
```

```python
import warnings
from lightgbm import LGBMClassifier
from sklearn.multiclass import OneVsRestClassifier

warnings.filterwarnings('ignore') 
alexander = -1
# Samaneh Kouchaki
# This file includes multi and single label RF classifiers
#In all classifiers we run 10 runs of 5-fold crossvalidation in each run 20 percent keeps as test set and 80 percent train and validation sets
#for setting the classifier threshold (it is based on best accuracy on validation set)
#10 fold cross validation and 5 runs
class EnsembleLearning_multilabel:
    #load data and label
    #filename1 is the feature matrix (rows:isolates and cols: features) and filename2 is labels one for each isolates
    def __init__(self, feat_df,label_df):
        self.feat=feat_df.values
        self.label=label_df.values 
        
    #multi-label RF
    # name1: all mutations list and name2: important features list for all drugs 
    #name1 and name2 are for loading important features. It is used only if not none otherwise the whole feature space is used
    def ensemble_multilabel(self,name1 = None , name2 = None):
        #keeping performance
        # 10 runs of 5 fold (50) for EMB, INH, RIF, PZA, XDR, MDR (6)
        accm=np.zeros((20,4))
        senm=np.zeros((20,4))
        spsm=np.zeros((20,4))
        aucm=np.zeros((20,4))
        kk = 0
        
        wdata = self.feat
        wlabel = self.label
        wlabel = wlabel.astype(int)
        
        inds =[]
        
        #load important features to be used for training
        if name1 != None:
            # it load
            muts = pd.read_csv(name1,header=None)
            feats = pd.read_csv(name2,header=None, names=list(range(200)))
            muts = muts.values
            feats = feats.values
            x = feats[0,:]
            for i in range(len(x)):
                if x[i][0] == "[" and x[i][-1] == "]":
                    x[i] = x[i][2:-2]
                    ii = i
                    break
                elif x[i][0] == "[":
                    x[i] = x[i][2:-1]
                elif x[i][-1] == "]":
                    x[i] = x[i][2:-2]
                    ii = i
                    break
                else:
                    x[i] = x[i][2:-1]
            x = x[:ii+1]
            inds = self.find_features_multi(muts,x)
            wdata = wdata[:,inds] # only keep important features

        
        for i in range(3):
            model = OneVsRestClassifier(LGBMClassifier())
            test_fold = np.zeros(len(wlabel))  
            #use iterative split
#             ind = iterative_split(wlabel, 5)
            ind = proba_mass_split(wlabel, 5)
            
            #split data based on iterative method
            for j in range(len(ind)):
                for k in range(len(ind[j])):
                    test_fold[ind[j][k]]=j
            ps = PredefinedSplit(test_fold)
            
            for trind, teind in ps.split():
                # 80% as train and validation
                tr=wdata[trind]
                trl=wlabel[trind]
                
                #20% as test
                te=wdata[teind]
                tel=wlabel[teind]
             
                # train validation split based on iterative CV
#                 tind = iterative_split(trl, 5)
                tind = proba_mass_split(trl, 5)
    
                test_label = trl[tind[0]]  
                test=tr[tind[0]]
                tind=np.concatenate(tind[1:5],axis=0)
                train_label=trl[tind]
                train=tr[tind]
                
                # fit the model 
                model.fit(train,train_label)
                
                # find treshold for each label
                pred=model.predict_proba(test)
                thr=[]
                for j in range(len(pred[0])):
                    thr.append(find_thr(pred[:,j],test_label[:,j]))
                pred=model.predict_proba(te)
                prr=np.zeros((len(tel),len(tel[0])))
                
                for j in range(len(pred[0])):
                    prr[:,j]= np.where(pred[:,j] > thr[j], 1, 0)
                
                #accuracy per label
                for j in range(len(prr[0])):
                    acc,sen,sps,auc=performance_calculation(tel[:,j],prr[:,j],prr[:,j])
                    accm[kk][j]=acc
                    senm[kk][j]=sen
                    spsm[kk][j]=sps
                    aucm[kk][j]=auc
                kk=kk+1
                print(kk)
        # return
        f = open("./dataset_pff/data_multi/lgb_res2.txt", "a")
        f.write("lgb_res ={'accm': np.array("+str((accm*100).tolist())+"),\n'senm':np.array(" + str((senm*100).tolist())+"),\n'spsm': np.array("+str((spsm*100).tolist())+ "),\n'aucm' : np.array("+str((aucm*100).tolist())+")}\n")
        f.close()
        print("accm: ", accm*100, "\nsenm: ", senm*100, "\nspsm: ",spsm*100, "\naucm:", aucm*100)
        return accm *100, senm*100, spsm*100, aucm*100
```
